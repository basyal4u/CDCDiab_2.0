# Processing Column of PFQ

```python
# PFQ051 - Limited in amount of work you can do
PFQ_reading1 = PFQ['PFQ051']
print('PFQ_reading1 shape = ', PFQ_reading1.shape)
print(PFQ_reading1.head())
print(PFQ_reading1.describe())
```

```python
# PFQ061C - Walking up ten steps difficulty
PFQ_reading2 = PFQ['PFQ061C']
print('PFQ_reading2 shape = ', PFQ_reading2.shape)
print(PFQ_reading2.head())
print(PFQ_reading2.describe())
```

```python
print('PFQ_reading1 missing values =  ', PFQ_reading1.isnull().sum(), 'from ', PFQ_reading1.shape[0], 'observations')
print('PFQ_reading2 missing values =  ', PFQ_reading2.isnull().sum(), 'from ', PFQ_reading2.shape[0], 'observations')
```

```python
All_PFQ = pd.concat([PFQ_reading1,PFQ_reading2],axis=1)
print('All_PFQ shape = ', All_PFQ.shape)
print(All_PFQ.head())
All_PFQ.describe()
```

```python
All_PFQ.to_csv('all_PFQ.csv')
```
